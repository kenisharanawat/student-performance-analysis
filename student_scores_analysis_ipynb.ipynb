```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.DataFrame({
    "StudyHours":[2,4,6,8],
    "Score":[50,65,75,90]
})
print(data.corr())
plt.plot(data["StudyHours"], data["Score"])
plt.title("Study Hours vs Score")
plt.xlabel("Study Hours")
plt.ylabel("Score")
plt.show()
```
